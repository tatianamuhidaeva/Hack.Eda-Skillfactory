```python
df4=df3.dropna()
df4[df4.rest_id=='9b73035bb90cd7de4b9c1ff01e44f178'].sort_values(by ='order_created_datetime')
```

```python
#нужно найти разницу во времени между двумя заказами, объединенными в один мультизаказ
# делим датасет с объединенными заказами на 2 датасета по значению first_in_multiorder_flg, а потом объединяем 
#два датасета, чтобы получить датасет с доп столбцом со временем парного заказа, чтобы можно было посчитать разницу
mask=df4['first_in_multiorder_flg'] == 1
mask
df5, df6 = df4[mask], df4[~mask]
df5
df6
df7=df5.merge(df6, left_on='batched_with_order_id', right_on='order_id', how ='inner')
df7

```

```python
df7.order_created_datetime_x=pd.to_datetime(df7.order_created_datetime_x)
df7.order_created_datetime_y=pd.to_datetime(df7.order_created_datetime_y)
df7
#находим разницу во времени между совместными заказами и сортируем по возрастанию. Видим, что минимальное время
#между совместными заказами 2 секунды, а максимальное 8 часов 39 минут (ошибка? выброс?), предыдущая дельта по
#времени между совместными 34 минуты
abs(df7.order_created_datetime_x - df7.order_created_datetime_y).sort_values()
```

```python
#при просмотре дельты времени между заказами, объединенными в 1 мультизаказ - обнаружился 1 случай, в котором разница между заказами составила 8 часов - это мог быть случай, когда человек сделал заказ к определенному времени (оформлен ночью, а желаемая дата доставки оказалась после открытия ресторана и совпала с размещенным вторым заказом)
#либо это ошибка
#один заказ оформили ночью в 4 утра к определенному времени, видимо ресторан открылся в 12, и второй заказ упал в 12:39 и поэтому их объединили
#предположим, что максимальное время между заказами 40 мин в рамках режима работы ресторан
df7.loc[168]
```

```python
#необходимо посчитать максимальную дельту по времени между 2мя заказами, объединенными в один мультизаказ, за
#исключением выброса  168 строка - получилось 34 минуты
df8=df7.drop([168])
df8.order_created_datetime_x=pd.to_datetime(df8.order_created_datetime_x)
df8.order_created_datetime_y=pd.to_datetime(df8.order_created_datetime_y)
abs(df8.order_created_datetime_x - df8.order_created_datetime_y).sort_values()
```

```python
#средняя дельта во времени между двумя объединенными заказами - 9 мин
abs(df7.order_created_datetime_x - df7.order_created_datetime_y).mean()
```

```python
#медианная дельта во времени между двумя объединенными заказами - 6 мин
abs(df7.order_created_datetime_x - df7.order_created_datetime_y).median()
```
