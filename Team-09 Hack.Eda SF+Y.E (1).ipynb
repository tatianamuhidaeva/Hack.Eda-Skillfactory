## Условие задачи

Имеется неизвестный нам алгоритм по объединению заказов и результат его работы на выборке ресторанов А. В задаче нужно предсказать поведение алгоритма на выборке ресторанов B. Скачать данные можно на странице с описанием задачи, в формате .xlsx. Это не привычный нам по курсу формат .csv, и смена расширения на .csv здесь не сработает. Поэтому первым делом вам нужно найти способ загрузить эти данные в Python. Рассмотрим по очереди, что означает каждый столбец.

**brand_name**. Это и есть выборка ресторанов (А или В). Все строки с brand_name == А - это реальные обезличенные данные о заказах в Яндекс Еде, при этом работал алгоритм, объединяющий некоторые заказы (не более двух!). Все строки с brand_name == В - это тоже реальные данные о заказах, но в других ресторанах, для которых алгоритм объдинения заказов не работал. Ваша задача - предсказать, какой процент заказов был бы объединен, если бы работал алгоритм объединения. Например, если у нас всего 3 заказа, и первый объединен со вторым, то объединено 66.7% заказов.

**rest_id**. Идентификатор конкретного ресторана (то есть конкретного адреса, откуда курьер забирает заказ). Заказы могут объединяться только если у них один и тот же rest_id.

**order_id**. Уникальный идентификатор заказа.

**batched_with_order_id**. Идентификатор заказа, объединенного с текущим (при наличии). Очевидно, что в выборке В это поле всегда пусто.

**first_in_multiorder_flg**. Является ли заказ первым в мультизаказе (при объединении двух заказов).

**courier_id**. Уникальный идентификатор курьера.

**order_items_cnt**. Количество позиций в заказе.

**city**. Уникальный идентификатор города. Все заказы выборки А принадлежат одному и тому же городу, а вот в выборке В есть разные города.

**order_created_datetime**. Время создания заказа. Обратите внимание, указан 2015 год, но это не значит, что задача старая. Дело в том, что данные обезличены, и все даты тоже изменены.

Выразим вероятность объединения заказов в зависимости от времени между ними для группы А.

```python
a.sort_values(['rest_id', 'order_created_datetime']).reset_index(drop=True)
```

```python
# добавляем уникальный id_мультизаказа: 'batch_id' (для дальнейшего удобства работы с мультизаказми)
def get_batch_id(order):
    """
    формируем batch_id как строку из двух IDs заказов: сначала максимальный. потом минимальный
    можно было бы через lambda, но так нагляднее
    """
    batch_id = order['batched_with_order_id']
    if order['batched_with_order_id'] is not None:
        if str(order['order_id']) > str(order['batched_with_order_id']):
            batch_id = str(order['order_id']) + '|' + str(order['batched_with_order_id'])
        else: batch_id = str(order['batched_with_order_id']) + '|' + str(order['order_id'])
    return batch_id

a['batch_id'] = a.apply(get_batch_id, axis=1)

display(a['batch_id'])

a_batched = a[a['batched_with_order_id'].notna()]
```

```python
a_batched.info()
```

```python
# a_batched только мультизаказы
```

```python
a_batched.head()
```

```python
batched_min = a_batched[['batch_id', 'order_created_datetime']].groupby(["batch_id"], as_index=False).min().reset_index()
batched_min['first_second'] = 'first'
batched_max = a_batched[['batch_id', 'order_created_datetime']].groupby(["batch_id"], as_index=False).max().reset_index()
batched_max['first_second'] = 'second'
batched_all = pd.concat([batched_min, batched_max])
batched_all
display(batched_all)
print(batched_all['first_second'].unique())
```

```python
 a = pd.merge(a, batched_all, how='left', on=['batch_id', 'order_created_datetime'])
```

```python
a.info()
```

```python
# теперь для каждого заказа вычислим время до пердыдущего заказа (с точностью до минут)
# DataSet _a_ отсортирован по rest_id и времени создания заказа
a = a.sort_values(['rest_id', 'order_created_datetime']).reset_index(drop=True)
orders_time = {'order_id':[],'time_shift':[]}
old_rest_id = None
old_order_time = None
for idx, row in a[['rest_id', 'order_id', 'order_created_datetime']].iterrows():
    if old_rest_id == row['rest_id']:
        time_shift = (row['order_created_datetime'] - old_order_time).seconds // 60
    else:
        old_rest_id = row['rest_id']
        time_shift = None
    old_order_time = row['order_created_datetime']
    orders_time['order_id'].append(row['order_id'])
    orders_time['time_shift'].append(time_shift)
print(len(orders_time))        

# конвертируем словарь в DataFrame
order_time_df = pd.DataFrame(orders_time)
# order_time_df['time_shift'].astype(int)
order_time_df.info()
order_time_df.head()
```

```python
# добавляем (join) time_shift к основному набору данных
a = pd.merge(a, order_time_df, how='left', on=['order_id'])
a.head()
```

```python
all_orders_by_timeshift = a[['time_shift', 'order_id']].groupby(["time_shift"], as_index=False).count().reset_index()
all_orders_by_timeshift = all_orders_by_timeshift.rename(columns={"order_id": "orders_count"})[['time_shift', 'orders_count']]
all_orders_by_timeshift
```

```python
second_orders_by_timeshift = a[['time_shift', 'order_id']][a['first_second'] == 'second'].groupby(["time_shift"], as_index=False).count().reset_index()
second_orders_by_timeshift = second_orders_by_timeshift.rename(columns={"order_id": "second_orders"})[['time_shift', 'second_orders']]
second_orders_by_timeshift
```

```python
a_by_timeshift_df = pd.merge(all_orders_by_timeshift, second_orders_by_timeshift, how = 'outer', on=['time_shift'])
a_by_timeshift_df['second_orders'] = a_by_timeshift_df['second_orders'].fillna(0)
a_by_timeshift_df
```

```python
def get_probability(ts_row):
    probability = 0
    if ts_row['orders_count'] is not None and ts_row['second_orders'] is not None:
        probability = ts_row['second_orders'] / ts_row['orders_count']
    return probability

a_by_timeshift_df['probability'] = a_by_timeshift_df.apply(get_probability, axis=1)
probability_by_timeshift = a_by_timeshift_df[['time_shift', 'probability']]
probability_by_timeshift.head(15)
# при увеличении временного интервала между заказами - вероятность объединения падает. 
```

```python
# Т.е. если второй заказ пришел через столько-то секунд после первого, 
# то вероятность их объединения probability_by_timeshift(time_shift)
```

```python
# теперь берем DataFrame с ресторанами B, для каждого заказа аналогично вычисляем время в минутах до предыдущего и 
# умножим это время на вероятность, просуммируем полученные значения, умножим на два (т.к. в мультизаказе два заказа) 
# и просуммируем - это будет количество заказов, доставляемых одновременно. 
# Разделим на общее количество заказов и получим искомую величину метрики
```

```python
# итак, есть DataFrame _B_ содержащий только рестораны В
b.info()
b['brand_name'].unique()
```

```python
# сортируем по ресторану и времени заказа.. 
b = b.sort_values(['rest_id', 'order_created_datetime']).reset_index(drop=True)
b.head()
```

```python
# вычисляем для каждого заказа время до предыдущего. 
orders_time = {'order_id':[],'time_shift':[]}
old_rest_id = None
old_order_time = None
for idx, row in b[['rest_id', 'order_id', 'order_created_datetime']].iterrows():
    if old_rest_id == row['rest_id']:
        time_shift = (row['order_created_datetime'] - old_order_time).seconds // 60
    else:
        old_rest_id = row['rest_id']
        time_shift = None
    old_order_time = row['order_created_datetime']
    orders_time['order_id'].append(row['order_id'])
    orders_time['time_shift'].append(time_shift)
print(len(orders_time))

# конвертируем словарь в DataFrame
order_time_df = pd.DataFrame(orders_time)
# order_time_df['time_shift'].astype(int)
order_time_df.info()
order_time_df.head()
```

```python
# добавляем (join) time_shift к основному набору данных
b = pd.merge(b, order_time_df, how='left', on=['order_id'])
b.head()
```

```python
b_with_probability = pd.merge(b, probability_by_timeshift, how='left', on=['time_shift'])
display(b_with_probability[b_with_probability.order_id == '50ee0573db2bc2270247e7d21a744d7e'])
b_forecast = b_with_probability[['order_created_datetime', 'city', 'rest_id', 'order_id', 'time_shift', 'probability']]
b_forecast['probability'] = b_forecast['probability'].fillna(0)
# умножаем на 2 т.к. один заказт добавляется к предыдущему и два заказа едут вместе, а не один
b_forecast['joined'] = b_forecast['probability'] * 2

b_forecast.head()
```

```python
# количетсво заказов в ресторанах _B_ 
b_orders_count = b_forecast['order_id'].count()
# количество объедененных заказов:
b_joined_count = int(b_forecast['joined'].sum())
print(f"количетсво заказов в ресторанах _B_: {b_orders_count}, прогноз количества объединенных заказов: {b_joined_count}")
print(f"\nИскомый прогноз % совмещения заказов в ресторанах _В_: {100*b_joined_count/b_orders_count:0.2f}%.")
```

Если объединять три заказа в один это даст дополнительно до 13 % в объединении заказов на выборке ресторанов В.

В алгоритм также необходимо заложить в явном виде ограничение на сумму количества продуктов в мультизаказе и если в первом заказе много продуктов, то во втором заказе (к нему) должно быть мало продуктов. 

Тогда, если в первом заказе много продуктов, то ресторан его долго делает и за это время поступает второй заказ и если во втором заказе мало продуктов ресторан успеет сделать оба заказа быстро, а курьер успеет донести за время меньше обещенного.
